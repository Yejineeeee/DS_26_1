## 1. Tuple (튜플)



*   한 번 할당되면 변경할 수 없는, **고정 길이**를 갖는 자료형
*   index는 **0부터 시작**한다.
*   튜플을 통째로 바꾸는 건 가능하지만, ***튜플을 하나씩 바꾸는 것은 불가능***하다.
*   ***'+ 연산자'*** (튜플 이어붙이기), ***'정수 곱하기'*** (여러 개의 튜플 복사본 생성) 기능을 지원한다.
*   **데이터 무결성 보장** (-> 튜플은 한 번 저장되면 값이 변경되지 않는다.)


```python
tup_1 = (4, 5, 6)
tup_2 = tuple([4, 5, 6])
tup_3 = tuple('string')

print(tup_1, tup_2, tup_3)
```

```python
# 인덱스는 0부터 시작
tup_1[0]
```

```python
# 튜플을 통째로 바꾸는 것은 가능하지만,
# 튜플 일부만 바꾸는 것은 불가능하다.

tup = tuple(['foo', [1, 2], True])
print(tup)
# tup[2] = False -> error

tup = tuple([1, 2, 3])
print(tup)
```

```python
# 튜플은 "칸 안에 뭐가 들어있는지"를 고정하는 거지,
# 그 안에 들어있는 객체가 스스로 변하는 건 막지 않는다.
# 리스트가 칸 안에 들어있으면, 그 리스트 자체는 가변 객체니까
# append나 del이 가능하다.

tup_2 = tuple(['foo', [1, 2], True])
tup_2[1].append(3)
print(tup_2)

del tup_2[1][1]
print(tup_2)
```



*   **리스트 슬라이싱**

```python
s = 'abcde'

s[1:4] # 'bcd' → 인덱스 1~3 (stop 미포함)
s[1:4:2] # 'bd'  → 인덱스 1~3에서 2칸 간격
s[:] # 'abcde' → 전체 복사
s[:3] # 'abc' → 처음~2
s[2:] # 'cde' → 2~끝
s[-3:-1] # 'cd'  → 음수 인덱스도 stop 미포함
```

## 2. list (리스트)



*   **append(값)** : 리스트 끝에 값 추가
*   **insert(위치, 값)** : 원하는 위치에 삽입
*   **pop(인덱스)** : 해당 위치 원소를 꺼내면서 제거
*   **remove(값)** : 값으로 찾아서 제거
*   **extend(리스트)** : 다른 리스트를 이어붙이기



```python
list_a = [2, 3, 5, None]
list_b = list(tup_dot)
print(list_a)
print(list_b)
list_b[1] = "yes"
print(list_b)

# 1. append(값)
list_a.append('red')
print(list_a)

# 2. insert(위치, 값)
list_b.insert(1, 'No')
print(list_b)

# 3. pop(인덱스)
list_a.pop(2)
print(list_a)

# 4. remove(값)
list_b.remove('No')
print(list_b)

# 5. extend(리스트)
list_a.extend(list_b)
print(list_a)
```

## 4. dict (딕셔너리)


**<딕셔너리의 키로 사용 가능한 타입>**


1.   정수
2.   문자열
3.   튜플   
→ 공통점 : **불변 객체**   
→ **리스트**는 가변 객체라 키로 사용할 수 없다.




```python
d1 = {"a": "some value", "b": [1,2,3,4]}
d1[7] = "an integer" # 정수 키 → OK
d1[(1,2,3)] = "tuple" # 튜플 키 → OK
# d1[[1,2,3]] = "list" # 리스트 키 → error!
```



*   **in / not in** : 키 존재 여부 확인
*   **del d[key]** : 키-값 쌍 삭제
*   **pop(key)** : 키로 꺼내면서 삭제
*   **key()** / **values()** / **items()** : 키-값 조회
*   **update()** : 병합 및 덮어쓰기



```python
#in, not in 문법으로 key 값 확인 가능
print("b" in d1)

d1[5] = "some value"
d1["dummy"] = "another value"

print(d1)

# del d[key]로 키-값 쌍 삭제 가능
del d1[5]
print(d1)

# pop(key)로 키로 꺼내면서 삭제 가능
ret = d1.pop("dummy")
print(ret)
print(d1)

# key() / values() / items()로 키-값 조회
# items()는 딕셔너리의 키와 값을 (키, 값) 튜플 쌍으로 묶어서 반환
print(list(d1.keys()))
print(list(d1.values()),'\n')
print(list(d1.items()),'\n')   #키값들이나 value값들을 뽑아와서 for문 등 반복 작업이 필요할 때

print(d1)
d1.update({"b":"foo", "c":12})  # 같은 key 값에 대해서는 overwirte 됨
print(d1)
```

## 5. function (함수)

```python
# 매우 기본적인 함수
def add(x, y):
    return x + y

add(42, 23)

# 기본 값을 가지는 함수
def add(x, y, z=0):
    return x + y + z

print(add(42, 23))
print(add(42, 23, 1024))

# 기본 값을 None으로 지정할 수도 있음
def add(x, y, z=None):
    if z != None:
        return x + y + z
    else:
        return x + y

print(add(42, 23))
print(add(42, 23, 1024), "\n")

#같은 함수의 약간 다른 표현
def add(x, y, z=None):
    if z: # None 일때만 False가 됨
        return x + y + z
    else:
        return x + y

print(add(42, 23))
print(add(42, 23, 1024))
```

```python
#함수 이름을 변수에 담아서, 다른 이름으로 호출할 수도 있음

def add(x, y):
    return x + y

def substract(x, y):
    return x - y

op = add # op refers to add()
print(op(42, 23))
op = substract # op refers to substract()
print(op(42, 23))
```

## 7. High order functions (고차 함수)



**고차 함수** : 함수를 인자로 받거나 함수를 결과로 돌려주는 함수

*   **map(func, iterable)** : 각 원소에 함수 적용; 모든 원소를 일괄적으로 변환해서 새로운 결과를 만든다.   
→ map()이 for문 보다 빠르다.
*   **filter(func, iterable)** : 조건에 맞는 원소만 걸러낸다.
*   **reduce(func, iterable)** : 원소를 묶어서 하나의 값으로 줄여나간다.





```python
vector = [1,2,3,4,5]
squared = []
for i in vector:
    squared.append(i**2)
print(squared)    ## 일반적인 코드 작성 방식

# 1. map()
def square(x):
    return x*x
print(list(map(square, vector)), '\n')

# 이를 응용한 high order function 한 줄 코딩
squared_2 = list(map(lambda x: x**2, vector))
print(squared_2)
```

```python
# 2. filter()
vector = [1,2,3,4,5]
list(filter(lambda x: x % 2 == 0, vector))

# 조건에 대한 값이 true인 것만 필터링
```

```python
# 3. reduce()
# list에서 순서대로 값들의 쌍에 대해서 연산을 해주는 역할

product = 1
list_1 = [1,2,3,4]
for num in list_1:
    product = product * num
print (product)
```

## 8. 데이터 Read/Write (File I/O)

```python
# 1. 파일 쓰기 (Write)
# 'with' 구문을 사용하면 파일을 다 쓴 후 자동으로 close()가 호출되어 안전함

data_to_save = [
    "1주차: 데이터 사이언스 개론\n",
    "2주차: 파이썬 기초 복습\n",
    "3주차: numpy 및 Pandas 데이터 분석\n"
]

# 'w' 모드: 쓰기 (기존 내용 삭제됨), 'a' 모드: 이어쓰기 (append)
# encoding='utf-8': 한글 깨짐 방지를 위해 encoding 지정
with open('curriculum.txt', 'w', encoding='utf-8') as f:
    for line in data_to_save:
        f.write(line)

print("파일 저장이 완료되었습니다. (curriculum.txt)")
```

```python
# 2. 파일 읽기 (Read)
# 저장된 파일을 불러와서 리스트로 변환

read_data = []

with open('curriculum.txt', 'r', encoding='utf-8') as f:
    lines = f.readlines()  # 모든 줄을 읽어서 리스트로 반환

    for line in lines:
        # strip(): 문자열 앞뒤의 공백이나 줄바꿈 기호(\n)를 제거
        clean_line = line.strip()
        read_data.append(clean_line)
        print("Read line: ",clean_line)

print("\n최종 데이터 리스트:", read_data)
```

## 10. Nested list comprehension (중첩 리스트 컴프리헨션)

```python
### 조금 더 복잡한 예시, x는 짝수만, y는 홀수만으로 짝 만들기
v = []
for x in range(5):
    if x % 2 == 0:
        for y in range(5):
            if y % 2:    ### y%2 == 1: 과 같은 표현, 파이썬은 숫자 1을 True로 인식하기 때문
                v.append((x,y))
print(v)

## 조금 더 코드의 가독성을 좋게 하려면,
v_2 = [(x,y) for x in range(5) if x % 2 == 0
             for y in range(5) if y % 2]
print(v_2)
```

## 11. enumerate, zip


*   **enumerate()** : 반복할 때 인덱스와 값을 동시에 돌려주는 함수; 인덱스와 섞어서 튜플을 만들어주는 함수
*   **zip()** : list 2개를 묶어서 tuple을 만들어주는 함수



```python
# enumerate()
import random

v1 = [random.randint(1, 100) for i in range(5)]
print(v1)
for val in enumerate(v1):
    print(val, type(val))
print('\n')
for i, v in enumerate(v1):
    print(i, v)
print("\n")

for i, v in enumerate(v1, start = 3):   #인덱스 시작 포인트도 변경 가능
    print(i, v)
```

```python
# 2. zip()
v1 = [random.randint(1, 100) for i in range(5)]
v2 = [random.randint(1, 100) for i in range(5)]
print('v1 = ', v1)
print('v2 = ', v2)
for i in zip(v1,v2):
    print(i, type(i))
print('\n')
for a, b in zip(v1,v2):
    print(a, b)
print(list(zip(v1,v2)))

# (키,값) 쌍으로 묶이므로 바로 딕셔너리로 만들 수 있다.
# 길이가 다르면, 짧은 쪽에 맞춘다.
```

## 12. dict comprehension

## 13. Label Encoding



*   문자열을 숫자로 매핑한다.
*   머신러닝에서 자주 쓰이는 패턴으로, 카테고리 문자열에 고유 번호를 붙여주는 것이다.



```python
categories = ["Red", "Green", "Blue", "Yellow"]

label_map = {name: i for i, name in enumerate(categories)}
print(label_map)
# {'Red': 0, 'Green': 1, 'Blue': 2, 'Yellow': 3}
```

## 15. List vs Dict 탐색 속도



*   **리스트**에서 in으로 검색하면 처음부터 끝까지 하나씩 확인해야 해서 느리다. (O(n))
*   반면 **딕셔너리**는 해시 테이블 기반이라 키를 바로 찾아가기 때문에 **사실상 즉시 검색**된다. (O(1))
*   데이터가 클수록 이 차이가 극적으로 벌어진다.


**Hash 값 확인; 왜 Dict가 빠른가?**


*   딕셔너리가 빠른 이유는 키를 해시값으로 변환해서 저장 위치를 바로 계산하기 때문이다.
*   문자열과 튜플은 불변이라 해시값을 만들 수 있지만, 리스트는 가변이라 해시가 불가능해서 키로 쓸 수 없다,



```python
key_str = "SINC Lab"
print(hash(key_str))       # 8562186250956213076

key_tup = (2026, "Spring")
print(hash(key_tup))       # -3019884025915201754

# key_list = [1,2,3]
# print(hash(key_list))      # error (리스트는 가변이라 해시 불가능)
```

## 16. Exercise (list comprehension + zip)

```python
# 원래 코드 (5줄)
names = ["Alice", "Bob", "Charlie"]
scores = [85, 90, 78]
result = []
for i in range(len(names)):
    if scores[i] >= 80:
        result.append(names[i])
# ['Alice', 'Bob']

# list comprehension + zip으로 한 줄
[n for n, s in zip(names, scores) if s >= 80]
# ['Alice', 'Bob']
```
